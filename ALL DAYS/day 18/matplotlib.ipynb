# MATPLOTLIB<hr/>
#### PYPLOT => It is a module of matplotlib which is used to plot graphs

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]


plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('simple line chart')
plt.show()
```

```python
import matplotlib.pyplot as plt

days = list(range(1,37))
tempreatures = [33,66,65,0,59,60,62,64,70,76,80,81,80,83,90,79,61,53,50,49,53,48,45,39,38,39,60,66,76,81,85,87,90,89,86,80]
# print(len(tempreatures))
plt.plot(days, tempreatures, marker='+', color='red')
plt.title("Daily temperatures", fontsize=24)
plt.xlabel("Days", fontsize=14)
plt.ylabel("Temperature (F)", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
genres = ['Comedy', 'Action', 'Thriller', 'Horror', 'Romance', 'Crime', 'Documentary']
book_sold = [12, 30, 15, 20, 25, 32, 18]

plt.bar(genres, book_sold, color=['green', 'blue', 'yellow','red', 'pink', 'black', 'brown'], edgecolor='y', linewidth=3.5, linestyle=':', width=0.8, alpha=0.6, label='Book Sold')
plt.ylabel('Book Sold')
plt.xlabel('Genres')
plt.show()
```

```python
import matplotlib.pyplot as plt
genres = ['Comedy', 'Action', 'Thriller', 'Horror', 'Romance', 'Crime', 'Documentary']
book_sold = [12, 30, 15, 20, 25, 32, 18]

plt.barh(genres, book_sold, color=["cyan", "blue"])
plt.ylabel('Book Sold')
plt.xlabel('Genres')
plt.show()
```
